# 深度学习模型训练和评估

```python
import os
import sys
import pandas as pd
import subprocess
```

```python
project_dir = '.'
script_path = os.path.join(project_dir, 'movie_box_office_analysis', 'src', 'deep_learning.py')
print('执行脚本:', script_path)

if os.path.exists(script_path):
    result = subprocess.run([sys.executable, script_path], cwd=project_dir, capture_output=True, text=True)
    print('脚本输出:')
    print(result.stdout)
    if result.stderr:
        print('错误信息:')
        print(result.stderr)
else:
    print('脚本不存在')
```

```python
# 查看models目录
models_dir = os.path.join(project_dir, 'movie_box_office_analysis', 'results', 'models')
print('models目录文件:')
if os.path.exists(models_dir):
    for f in os.listdir(models_dir):
        print(f'  {f}')
else:
    print('  models目录不存在')
```

```python
# 查看charts目录
charts_dir = os.path.join(project_dir, 'movie_box_office_analysis', 'results', 'charts')
print('charts目录文件:')
if os.path.exists(charts_dir):
    for f in os.listdir(charts_dir):
        print(f'  {f}')
else:
    print('  charts目录不存在')
```

```python
# 加载性能数据
perf_file = os.path.join(project_dir, 'movie_box_office_analysis', 'results', 'deep_learning_performance.csv')
if os.path.exists(perf_file):
    df = pd.read_csv(perf_file)
    print('性能数据:')
    print(df)
else:
    print('性能数据文件不存在')
```
