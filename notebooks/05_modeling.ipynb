本笔记本用于执行传统机器学习建模，包括模型训练、评估和比较等操作。

```python
# 导入必要的库
import os
import sys
import pandas as pd
```

## 1. 检查项目结构

```python
# 检查当前工作目录
print(f'当前工作目录: {os.getcwd()}')

# 检查项目结构
project_dir = '..'
print('项目结构:')
for root, dirs, files in os.walk(project_dir):
    level = root.replace(project_dir, '').count(os.sep)
    indent = ' ' * 2 * level
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 2 * (level + 1)
    for file in files[:5]:  # 只显示前5个文件
        print(f'{subindent}{file}')
    if len(files) > 5:
        print(f'{subindent}... 等 {len(files) - 5} 个文件')
```

## 2. 执行传统机器学习建模

```python
# 执行传统机器学习建模
print('正在执行传统机器学习建模...')

# 导入传统建模模块
sys.path.insert(0, project_dir)
from src.modeling import TraditionalModeling

# 初始化传统建模对象
traditional_modeler = TraditionalModeling()

# 执行完整的传统建模流程
traditional_modeler.run()

print('传统机器学习建模完成！')
```

## 3. 查看模型结果

```python
# 查看模型文件
models_dir = os.path.join(project_dir, 'results', 'models')
if os.path.exists(models_dir):
    print('生成的模型文件:')
    for file in os.listdir(models_dir):
        if file.endswith('.joblib') or file.endswith('.pkl'):
            print(f'  - {file}')
else:
    print('模型目录不存在')

# 查看模型比较报告
reports_dir = os.path.join(project_dir, 'results', 'reports')
model_comparison_path = os.path.join(reports_dir, 'model_comparison_metrics.csv')
if os.path.exists(model_comparison_path):
    model_comparison = pd.read_csv(model_comparison_path)
    print('\n模型比较指标:')
    display(model_comparison)
else:
    print('模型比较报告不存在')
```
